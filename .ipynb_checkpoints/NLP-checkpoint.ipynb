```python
import nltk
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
```

Here we have tokenized our paragraph into sentences

```python
## tokenization-- converts paragraph-sentences-words
nltk.download('punkt_tab')
sentences = nltk.sent_tokenize(paragraph)
```

```python
print(sentences)
```

```python
stemmer = PorterStemmer()
```

```python
stemmer.stem("going")
```

```python
stemmer.stem("history")
```

```python
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
lemmatizer.lemmatize("drinking")
```

First we need to clean our paragraph

```python
import re
```

```python
corpus = []
for i in range (len(sentences)):
    review = re.sub('[^a-zA-Z]',' ',sentences[i])      #Here we have removed special characters from our tokenized paragraph(i.e sentences)
    review = review.lower()
    corpus.append(review)
```

```python
corpus
```

```python
nltk.download('stopwords')
```

```python
stopwords.words("english")                        #This command will give us all the english stopwods
```

Applying stemming on our corpus list

```python
##Stemming
for i in corpus:
    words = nltk.word_tokenize(i)            #This tokenizes our sentences into words
    for word in words:
        if word not in set(stopwords.words("english")):
            print(stemmer.stem(word))
```

```python
##Lemmatization
for i in corpus:
    words = nltk.word_tokenize(i)            #This tokenizes our sentences into words
    for word in words:
        if word not in set(stopwords.words("english")):
            print(lemmatizer.lemmatize(word))
```

```python
##Apply stopwords
import re
corpus = []
for i in range(len(sentences)):
    review = re.sub('[^a-zA-Z]',' ',sentences[i])
    review = review.lower()
    review = review.split()
    review = [lemmatizer.lemmatize(word) for word in review if not word in set(stopwords.words("english"))]
    review = ' '.join(review)
    corpus.append(review)
```

Bag of Words

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
```

```python
x=cv.fit_transform(corpus)
```

```python
cv.vocabulary_                      #This will give us the unique words with their index
```

```python
x[0].toarray()                        #This will give us the vectorized words for our first sentence in corpus
```

```python
cv3 = CountVectorizer(binary=True,ngram_range=(3,3))                #Here we are dividing into unique n-grams of size 3
x1 = cv3.fit_transform(corpus)
```

```python
cv3.vocabulary_
```

```python
cv4 = CountVectorizer(binary=True,ngram_range=(2,3))                #Here we are dividing into unique n-grams of size 2 and 3
x2 = cv4.fit_transform(corpus)
```
